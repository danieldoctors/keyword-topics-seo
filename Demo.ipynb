```python
url = 'https://www.danieldoctor.com/conferencista-de-marketing-digital/'
```

```python
# Importing web data
import requests
from bs4 import BeautifulSoup
# Language Management 
import nltk, re, string, collections
from nltk.util import ngrams
from nltk.stem import WordNetLemmatizer
#Graphics
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
%matplotlib inline
```

```python
#requests.encoding = 'ISO-8859-1'
text = requests.get(url)
soup = BeautifulSoup(text.content, features="html.parser")
text.content
```

```python
for script in soup(["script", "style"]):
    script.extract()
```

```python
document = soup.get_text()
lines = (line.strip() for line in document.splitlines())
chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
final_document = ' '.join(chunk for chunk in chunks if chunk)
final_document = final_document.lower()
words = final_document.split()
```

```python
final_document
```

```python
tokenized = final_document.split()
Ngrams = ngrams(tokenized, 3)
NgramsFreq = collections.Counter(Ngrams)

# what are the ten most popular ngrams in this Spanish corpus?
NgramsFreq.most_common(10)
```

```python
from gensim.corpora import Dictionary
from gensim.parsing.porter import PorterStemmer
from gensim.utils import simple_preprocess
from gensim.models import LdaMulticore
stem = PorterStemmer()
def lemmatize_stemming(text):
    return stem.stem(WordNetLemmatizer().lemmatize(text, pos='v'))
tokenized
```

```python
stop = []
def preprocess(text):
    result=[]
    for token in simple_preprocess(text) :
        if token not in stop and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

```python
processed = preprocess(final_document)
```

```python
dictionary = Dictionary([processed])
```

```python
dictionary
bow_corpus = [dictionary.doc2bow(doc) for doc in [processed]]
bow_corpus
lda_model =  LdaMulticore(bow_corpus, 
                                   num_topics = 5, 
                                   id2word = dictionary,                                    
                                   passes = 10,
                                   workers = 2)
```

```python
document_num = 0
bow_doc_x = bow_corpus[document_num]

for i in range(len(bow_doc_x)):
    print("Palabra {} (\"{}\") aparece {} veces.".format(bow_doc_x[i][0], 
                                                     dictionary[bow_doc_x[i][0]], 
                                                     bow_doc_x[i][1]))
```

```python
for idx, topic in lda_model.print_topics(5):
    print("Tema: {} \nWords: {}".format(idx, topic ))
    print("\n")
```
